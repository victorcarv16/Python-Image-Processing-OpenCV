```python
class Parametros:
  def __init__(self,f,a,b,eps):
    self.f = f
    self.a = a
    self.b = b
    self.eps = eps



import math
import pandas as pd
alpha = 0.618 # (-1+math.sqrt(5))/2

def razao_aurea(p):
    cols = ['a','b','lambda','mi','f(lambda)','f(mi)']    
    l = p.b - p.a
    a = p.a
    b = p.b
    y = a + (1-alpha)*(b-a)
    m = a + alpha*(b-a)
    
    df = pd.DataFrame([[a,b,y,m,p.f(y),p.f(m)]], columns=cols)
    
    while (l > p.eps):
       fy = p.f(y)
       fm = p.f(m)
          
       if (fy > fm):
       #Passo 2
           a = y
           y = m
           m = a + alpha*(b-a)
       else:  
       #Passo 3
           b = m
           m = y
           y = a + (1-alpha)*(b-a)

       l = b - a
    
       row = pd.DataFrame([[a,b,y,m,p.f(y),p.f(m)]],columns=cols)
       df = df.append(row, ignore_index=True)

    return  df
```
